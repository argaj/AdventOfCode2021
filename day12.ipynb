```python
import networkx as nx
from collections import Counter
```

```python
data = nx.read_edgelist('day12.txt', delimiter='-')
```

```python
nx.draw(data)
```

```python
Q = [['start']]
paths = []
while Q:
    path = Q.pop(0)
    if path[0] == 'start' and path[-1] == 'end':
        paths.append(path)
    else:
        for node in data.neighbors(path[-1]):
            if node.isupper() or (node.islower() and node != 'start' and path.count(node) < 1):
                Q.append(path + [node])
```

```python
len(paths)
```

```python
Q = [['start']]
paths = []
while Q:
    path = Q.pop(0)
    if path[0] == 'start' and path[-1] == 'end':
        paths.append(path)
    else:
        for node in data.neighbors(path[-1]):
            counts = [cnt for n, cnt in Counter(path + [node]).items() if n.islower()]
            if node.isupper() or (node.islower() and node != 'start' and counts.count(2) < 2 and all([cnt < 3 for cnt in counts])):
                Q.append(path + [node])
```
